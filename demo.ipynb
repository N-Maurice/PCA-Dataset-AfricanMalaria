```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Loading the numeric dataset from its source
numeric_columns = [1] + list(range(3, 26))

data = np.genfromtxt(
    'DatasetAfricaMalaria.csv',
    delimiter=',',
    skip_header=1,
    usecols=numeric_columns,
    dtype=float,
    encoding='utf-8',
    quotechar='"'
)

# Print the loaded numeric data shape
print(data.shape)

# View the first 5 rows
print(data[:5])
```

```python
# Check the loaded dataset
print(data.shape)

print(data[:5])
```

```python
# Handle missing values since there are many NaNs values
for i in range(data.shape[1]):
    mean_value = 0 if np.isnan(data[:, i]).all() else np.nanmean(data[:, i])

    data[:, i] = np.where(
        np.isnan(data[:, i]),
        mean_value,
        data[:, i]
    )

    print(mean_value)
```
